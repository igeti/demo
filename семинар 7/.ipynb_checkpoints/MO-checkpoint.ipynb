### Часть 1: Реализация методов

Полезная функция: plt.contour
Для всех экспериментов подберите параметры так, чтобы метод сошелся к ближайшему локальному минимуму. Все методы следует запускать из одной и той же точки.

<i> 1.1 Реализуйте методы GD, Momentum, NAG, Adagrad, Adadelta, Adam.</i>

```python
def GD(func, grad, X, lr=1e-3, eps=1e-5):
    fs = []
    gs = []
    points = []
    while np.abs(np.linalg.norm(grad(X))) > eps:
        X = X - lr * grad(X)
        points.append(X)
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs, points
```

```python
def Momentum(func, grad, X, lr=1e-3, eps = 1e-5, gamma = 0.9):
    fs = []
    gs = []
    points = []
    nu = np.ones(X.shape)*eps
    while np.abs(np.linalg.norm(grad(X))) > eps:
        nu = nu*gamma + lr * grad(X)
        X = X - nu
        points.append(X)
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs, points
```

```python
def NAG(func, grad, X, lr=1e-3, eps = 1e-5, gamma = 0.9):
    fs = []
    gs = []
    points = []
    nu = np.ones(X.shape)*eps
    while np.abs(np.linalg.norm(grad(X))) > eps:
        nu = nu*gamma + lr*grad(X - gamma*nu)
        X = X - nu
        points.append(X)
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs, points
```

```python
def Adagrad(func, grad, X, lr=0.1, eps=1e-5):
    fs = []
    gs = []
    points = []
    grad_sum = eps * np.ones(X.shape)
    while np.abs(np.linalg.norm(grad(X))) > eps:
        grad_sum += np.square(grad(X)+eps)
        X = X - lr * (np.divide(grad(X), np.sqrt(grad_sum)))
        points.append(X)
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs, points
```

```python
def Adadelta(func, grad, X, eps = 1e-5, gamma = 0.9):
    fs = []
    gs = []
    points = []
    nge = eps * np.ones(X.shape)
    pge = eps * np.ones(X.shape)
    dX = eps * np.ones(X.shape)
    while np.abs(np.linalg.norm(grad(X))) > eps:
        nge = gamma * nge + (1 - gamma)* np.square(grad(X))
        rmn = np.sqrt(nge)
        pge = gamma * pge + (1 - gamma) * dX
        rmp = np.sqrt(abs(pge))
        dX = np.divide(rmp, rmn+eps)
        dX = np.multiply(dX, grad(X))
        X = X - dX
        points.append(X)
        fs.append(func(X))
        gs.append(grad(X))    
    return fs, gs, points
```

```python
def Adam(func, grad, X, lr =0.1, betta1 = 0.9, betta2 = 0.999, eps = 1e-8):
    fs = []
    gs = []
    points = []
    m = eps * np.ones(X.shape)
    nu = eps * np.ones(X.shape)
    t = 0
    while np.abs(np.linalg.norm(grad(X))) > eps:
        t +=1
        m = betta1*m +(1 - betta1)*grad(X)
        nu = betta2*nu +(1 - betta2)*np.square(grad(X))
        m_ = m/(1-betta1**t)
        nu_ = nu/(1-betta2**t)
        nu_ = np.sqrt(nu_)+eps
        X = X - lr*np.divide(m_, nu_)
        points.append(X)
        fs.append(func(X))
        gs.append(grad(X))    
    return fs, gs, points
```

<i> 1.2 Сравните эти методы на функции $J(x, y) = x^2+y^2$</i>

```python
X = 10 * np.array(np.random.random_sample(2))
```

```python
print X
```

```python
f = lambda X: X[0]**2+X[1]**2
df = lambda X: np.array([2*X[0], 2*X[1]])
```

```python
def plot_(points,func, label):    
    x = np.arange(-10, 10, 0.5)
    y = np.arange(-10, 10, 0.5)
    X, Y = np.meshgrid(x, y)

    Z = func([X,Y])
    points = np.array(points)
    plt.figure(figsize=[5,5])
    CS = plt.contour(X, Y, Z)
    plt.scatter(points[::10,0],points[::10,1])
    plt.clabel(CS, inline=1, fontsize=10)
    plt.grid()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(label)
```

```python
function, gradient, points  = GD(f, df, X)
plot_(points, label='GD', func = f)

function, gradient, points  = Momentum(f, df, X)
plot_(points, label='Momentum',func = f)

function, gradient, points  = NAG(f, df, X)
plot_(points, label='NAG',func = f)

function, gradient, points  = Adagrad(f, df, X)
plot_(points, label='Adagrad',func = f)

function, gradient, points  = Adadelta(f, df, X)
plot_(points, label='Adadelta',func = f)

function, gradient, points  = Adam(f, df, X)
plot_(points, label='Adam',func = f)
```

###  Часть 2: Обучение нейронной сети

<i> 2.1 Сравните графики обучения для полносвязной нейросети на методах Adam, NAG и SGD (на MNIST). </i>

```python
#Parameters:
_LR = 0.01
_BATCH_SIZE = 32
_EPOCH = 1
_EPS=1.00000e-02
_RAND_STATE = 1
```

```python
#Load MNIST
import torchvision.transforms as transforms
import torchvision.datasets as dset
from torch.utils.data import DataLoader

to_numpy = lambda x: x.numpy()
transform = transforms.Compose([
                       transforms.ToTensor(),
                       transforms.Normalize((0.1307,), (0.3081,))
                    ])
train_dataset = dset.MNIST('.', train=True, download=True, transform=transforms.ToTensor())
test_dataset = dset.MNIST('.', train=False, transform=transforms.ToTensor())

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
class network1(nn.Module):

    def __init__(self):
        super(network1, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Linear(784, 100),
            nn.Sigmoid()
        )
        self.layer2 = nn.Sequential(
            nn.Linear(100, _BATCH_SIZE),
            nn.Sigmoid()
        )
        self.fc = nn.Linear(_BATCH_SIZE,10)

    def forward(self, x):
        
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.fc(out)
        return out
```

```python
#Where is NAG?
loss_func = nn.CrossEntropyLoss()
```

# Adam

```python
nn1 = network1()
optimizer1 = torch.optim.Adam(nn1.parameters(), lr=_LR)
error_train1 =[]
loss_train1 = []
for i in tqdm(range(_EPOCH)):
    for j,[image,label] in enumerate(train_loader):
        image = Variable(image)
        label = Variable(label)
        image = image.view(-1, 28*28)
        optimizer1.zero_grad()
        result = nn1.forward(image)
        loss = loss_func(result,label)
        loss.backward()
        optimizer1.step()
        loss_train1.append(loss.data.numpy()[0])
        #print loss.data.numpy()[0]
        if (loss.data.numpy()[0]<_EPS):
            break
    error_train1.append(np.array(loss_train1))
```

# SGD

```python
nn2 = network1()
optimizer2 = torch.optim.SGD(nn2.parameters(), lr=_LR)
error_train2 =[]
loss_train2 = []
for i in tqdm(range(_EPOCH)):
    for j,[image,label] in enumerate(train_loader):
        image = Variable(image)
        label = Variable(label)
        image = image.view(-1, 28*28)
        optimizer2.zero_grad()
        result = nn2.forward(image)
        loss = loss_func(result,label)
        loss.backward()
        optimizer2.step()
        loss_train2.append(loss.data.numpy()[0])
        if (loss.data.numpy()[0]<_EPS):
            break
    error_train2.append(np.array(loss_train2))
```

```python
print np.array(error_train2).ravel()
```

```python
plt.figure()
plt.xlabel(u"Номер итерации(N)")
plt.ylabel(u"CrossEntropyLoss")
plt.grid()
plt.plot(np.array(error_train1).ravel(), label = "Adam")
plt.plot(np.array(error_train2).ravel(), label = "SGD")
plt.legend()
```

<i> 2.2 Сравните графики обучения для сверточной нейросети на методах Adam, NAG и SGD. </i>

```python
class CNN(torch.nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.layer1 = torch.nn.Sequential(nn.Conv2d(1, 8, 3), nn.ReLU(), nn.MaxPool2d(2))
        self.layer2 = torch.nn.Sequential(nn.Conv2d(8, 16, 3), nn.ReLU())
        self.layer3 = torch.nn.Sequential(nn.Linear((11)*(11)*16, 10))
        
    def forward(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = x.view(_BATCH_SIZE, -1)
        x = self.layer3(x)
        return x
```

```python
_EPOCH =15
```

# Adam

```python
model1 = CNN()
optimizer1 = torch.optim.Adam(model1.parameters(), lr=_LR)
error_train1 =[]
loss_train1 = []
for i in tqdm(range(0,_EPOCH)):
    for j,[image,label] in enumerate(train_loader):
        x = Variable(image)
        y_= Variable(label)
        
        optimizer1.zero_grad()
        output = model1.forward(x)
        loss = loss_func(output,y_)
        loss.backward()
        optimizer1.step()
        if (loss.data.numpy()[0]<_EPS):
            break
        
        loss_train1.append(loss.data.numpy()[0])
    error_train1.append(np.array(loss_train1).mean())
```

# SGD

```python
model2 = CNN()
optimizer2 = torch.optim.SGD(model2.parameters(), lr=_LR)
error_train2 =[]
loss_train2 = []
for i in tqdm(range(0,_EPOCH)):
    for j,[image,label] in enumerate(train_loader):
        x = Variable(image)
        y_= Variable(label)
        
        optimizer2.zero_grad()
        output = model2.forward(x)
        loss = loss_func(output,y_)
        loss.backward()
        optimizer2.step()
        if (loss.data.numpy()[0]<_EPS):
            break
        
        loss_train2.append(loss.data.numpy()[0])
    error_train2.append(np.array(loss_train2).mean())
```

```python
plt.figure()
plt.xlabel(u"Номер эпохи(N)")
plt.ylabel(u"CrossEntropyLoss")
plt.grid()
plt.plot(np.array(error_train1), label = "Adam")
plt.plot(np.array(error_train2), label = "SGD")
plt.legend()
```
