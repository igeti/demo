```python
def pca(num_components, idx):
    model1 = PCA(n_components=num_components)
    model1.fit(X)
    embeding = model1.transform(X)
    recorstruction = model1.inverse_transform(embeding)
    error = reconstruction - X
    fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(16,4))
    ax1.set_title('original')
    ax1.imshow((X)[idx].reshape(64,64), cmap='gray')
    ax2.set_title('reconstruction')
    ax2.imshow((recorstruction)[idx].reshape(64,64), cmap='gray') 
    ax3.set_title('Error')
    plt.imshow((error)[idx].reshape(64,64), cmap='gray')
    plt.colorbar()
    plt.show()
```

```python
for i in xrange(1,201,50):
    pca(i,10)
    print('Number of components: {}'. format(i))
```

```python
def plot_dataset(dataset, x, y, margin=50, frame=2000., imsize=64):
   x1 = np.array((x - x.min()) * (frame / (x.max() - x.min())) + margin, dtype = np.int)
   y1 = np.array((y - y.min()) * (frame / (y.max() - y.min())) + margin, dtype = np.int)
   image = np.zeros((int(x1.max() + margin + imsize), int(y1.max() + margin + imsize))) + 1
   print 'Image size is %dx%d” % image.shape'
   for k in xrange(dataset.images.shape[0]):
       image[x1[k]:x1[k]+64, y1[k]:y1[k]+64] = dataset.images[k]
   plt.figure(figsize=(20, 20))
   plt.imshow(image, cmap='Greys_r')
   plt.axis('off')
   plt.show()
```

###  Часть 2: Автокодировщик

<i> 2.1 Обучите автокодировщик с архитектурой 4096 -> 200 -> 10 -> 200 -> 4096. В качестве нелинейностей используйте сигмоиду.</i>

```python
from torch.utils.data import DataLoader, Dataset
```

```python
class FacesDataset(Dataset):
    """Face Landmarks dataset."""

    def __init__(self, X,transform=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        
        self.X = X
       

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        
        return X[idx]
```

```python
dataset = FacesDataset(X)
```

```python
dataloader = DataLoader(dataset, batch_size =16, shuffle ='true')
```

```python
def fully_connected(arc, last_nonlinearity=True):
    layers = []
    for i in range(len(arc)-1):
        layers.append(nn.Linear(arc[i], arc[i+1]))
        if i != len(arc)-2 or last_nonlinearity:
            layers.append(nn.LeakyReLU())
    return nn.Sequential(*layers)
```

```python
fully_connected([4096,200,10],last_nonlinearity = False)
```

```python
class AE(nn.Module):
    def __init__(self,arc_encoder,arc_decoder):
        super(AE,self).__init__()
        
        for i in range(len(arc_encoder)):
            self.encoder = fully_connected(arc_encoder)
            self.decoder = fully_connected(arc_decoder)
    def forward(self,X):
        return self.decoder(self.encoder(X))
```

```python
model = AE([4096, 200, 10], [10,200,4096])
loss = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr = 0.001)

for epoch in xrange(10):
    losses =[]
    for x in dataloader:
        x_batch = Variable(x)
        y_batch = Variable(x, requires_grad = False)
        model.zero_grad()
        x_reconstruction = model(x_batch)
        loss_batch = loss(x_reconstruction, y_batch)
        loss_batch.backward()
        optimizer.step()
        losses.append(loss_batch.data[0])
    sys.stdout.write('\r {}'.format(np.mean(losses)))
```

```python
embedding = model.encoder(Variable(torch.Tensor(X))).data.numpy()
```

```python
reconstructions = model(Variable(torch.Tensor(X))).data.numpy()
```

```python
plt.imshow(reconstructions[0].reshape(64,64), cmap = plt.cm.Greys_r)
```
